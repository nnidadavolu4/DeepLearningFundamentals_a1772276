```python
import pandas as pd
import os
import numpy as np
from sklearn.model_selection import train_test_split

from google.colab import drive

drive.mount('/content/drive')


AMZN = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Nims - DL/Assignment3/AMZN_YTD.csv', index_col='Date', parse_dates=['Date'])
AFRM = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Nims - DL/Assignment3/AFRM_YTD.csv')

AMZN['Open'] = AMZN['Open'].str.replace('$', '', regex=False)
AMZN['High'] = AMZN['High'].str.replace('$', '', regex=False)
AMZN['Low'] = AMZN['Low'].str.replace('$', '', regex=False)
AMZN['Close/Last'] = AMZN['Close/Last'].str.replace('$', '', regex=False)

AMZN['Open'] = pd.to_numeric(AMZN['Open'])
AMZN['High'] = pd.to_numeric(AMZN['High'])
AMZN['Low'] = pd.to_numeric(AMZN['Low'])
AMZN['Close/Last'] = pd.to_numeric(AMZN['Close/Last'])
AMZN['Vol'] = pd.to_numeric(AMZN['Vol'])
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

# Load the data
data = AMZN['Close/Last'].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import SimpleRNN, LSTM, GRU, Dense
import matplotlib.pyplot as plt
from keras.optimizers import Adam


# Create sequences
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:(i + seq_length), 0])
        y.append(data[i + seq_length, 0])
    return np.array(X), np.array(y)

sequence_length = 10
X, y = create_sequences(scaled_data, sequence_length)

# Split the data
train_size = int(len(X) * 0.7)
val_size = int(len(X) * 0.15)
X_train, y_train = X[:train_size], y[:train_size]
X_val, y_val = X[train_size:train_size+val_size], y[train_size:train_size+val_size]
X_test, y_test = X[train_size+val_size:], y[train_size+val_size:]

# Reshape input for RNN [samples, time steps, features]
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_val = X_val.reshape((X_val.shape[0], X_val.shape[1], 1))
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))

# Custom accuracy function
def custom_accuracy(y_true, y_pred, threshold=0.05):
    within_threshold = np.abs(y_true - y_pred) <= threshold * y_true
    return np.mean(within_threshold)

# Evaluate the model
def evaluate_model(model, X, y):
    y_pred = model.predict(X)
    mse = mean_squared_error(y, y_pred)
    accuracy = custom_accuracy(y, y_pred)
    return mse, accuracy

# Build and train model function
def build_and_train_model(model_type):
    if model_type == 'RNN':
        model = Sequential([
            SimpleRNN(50, activation='relu', input_shape=(sequence_length, 1), return_sequences=True),
            SimpleRNN(50, activation='relu'),
            Dense(1)
        ])
    elif model_type == 'LSTM':
        model = Sequential([
            LSTM(50, activation='relu', input_shape=(sequence_length, 1), return_sequences=True),
            LSTM(50, activation='relu'),
            Dense(1)
        ])
    elif model_type == 'GRU':
        model = Sequential([
            GRU(50, activation='relu', input_shape=(sequence_length, 1), return_sequences=True),
            GRU(50, activation='relu'),
            Dense(1)
        ])

    optimizer = Adam(learning_rate=0.01)
    model.compile(optimizer=optimizer, loss='mse', metrics=['accuracy'])
    history = model.fit(X_train, y_train, epochs=150, batch_size=32,
                        validation_data=(X_val, y_val), verbose=0)

    return model, history

# Train and evaluate models
models = ['RNN', 'LSTM', 'GRU']
results = {}

for model_type in models:
    model, history = build_and_train_model(model_type)
    train_mse, train_acc = evaluate_model(model, X_train, y_train)
    val_mse, val_acc = evaluate_model(model, X_val, y_val)
    test_mse, test_acc = evaluate_model(model, X_test, y_test)

    results[model_type] = {
        'model': model,
        'history': history,
        'train_mse': train_mse,
        'train_acc': train_acc,
        'val_mse': val_mse,
        'val_acc': val_acc,
        'test_mse': test_mse,
        'test_acc': test_acc
    }

    print(f"\n{model_type} Results:")
    print(f"Train MSE: {train_mse:.4f}, Accuracy: {train_acc:.4f}")
    print(f"Validation MSE: {val_mse:.4f}, Accuracy: {val_acc:.4f}")
    print(f"Test MSE: {test_mse:.4f}, Accuracy: {test_acc:.4f}")

# Plot MSE values
plt.figure(figsize=(18, 6))
for i, model_type in enumerate(models, 1):
    plt.subplot(1, 3, i)
    plt.plot(results[model_type]['history'].history['loss'], label=f'{model_type} Train MSE')
    plt.plot(results[model_type]['history'].history['val_loss'], label=f'{model_type} Validation MSE')
    plt.axhline(y=results[model_type]['test_mse'], color='r', linestyle='--', label=f'{model_type} Test MSE')
    plt.title(f'{model_type} MSE Values')
    plt.xlabel('Epoch')
    plt.ylabel('MSE')
    plt.legend()

plt.tight_layout()
plt.show()

# Predict next 5 days using all models
def predict_next_5_days(model, last_sequence):
    predictions = []
    for _ in range(5):
        next_pred = model.predict(last_sequence.reshape(1, sequence_length, 1))
        predictions.append(next_pred[0, 0])
        last_sequence = np.roll(last_sequence, -1)
        last_sequence[-1] = next_pred
    return scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

last_sequence = scaled_data[-sequence_length:]
actual_values = scaler.inverse_transform(y_test[:5].reshape(-1, 1))

print("\nPredictions for the next 5 days:")
for model_type in models:
    predictions = predict_next_5_days(results[model_type]['model'], last_sequence)
    print(f"\n{model_type} Predictions:")
    for i in range(5):
        print(f"Day {i+1}: Actual {actual_values[i][0]:.2f}, Predicted {predictions[i][0]:.2f}")
```
